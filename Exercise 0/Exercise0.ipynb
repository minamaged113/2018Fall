### 1.3. Import packages in your code
* Import a package using ```import packageName```, then you can call function provided by the package by ```packageName.functionName()```
  * EXAMPLE:
      ```
      import numpy
      myEmptyArray = numpy.zeros([nRow, cCol], dtype=float)
      ```
* Import a package and give it another alias by ```import packageName as alias```, then you can use the alias to represent the package in your later code.
  * EXAMPLE:
      ```
      import numpy as np
      myEmptyArray = np.zeros([nRow, cCol], dtype=float)
      ```
* Directly import a function from a package by ```from packageName import functionName```, then you can directly call the function in your later code.
  * EXAMPLE:
      ```
      from numpy import zeros
      myEmptyArray = zeros([nRow, cCol], dtype=float)
      ```

* Example: we have a data file *myDataFile.mat* which contains following data object (or dictionaries): *dic1*, *dic2*, ... Then
```
import scipy.sio as sio
myDic1Data = sio.loadmat('myDataFile.mat')['dic1']
```
